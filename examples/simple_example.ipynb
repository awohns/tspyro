# Simple example of tspyro inference

We use a tree sequence outputted from `spatial_model_nearest_neighbor.txt`. This runs a forwards-time, spatial simulation from http://benhaller.com/slim/SLiM_Manual.pdf, section 15.5.

This simulation gets the coordinates of each parent, adds a normal draw to each (thus deviating the offspringâ€™s x and y coordinates from those of the parent), asks the subpopulation to clamp the new position into bounds, and sets the final position into the child. The bounds are [0,1].

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error

import pyro
import torch
import tskit
import pyslim
import sys
sys.path.insert(1, "../")
import tspyro

import matplotlib.pyplot as plt
from IPython.display import SVG
from matplotlib.animation import FuncAnimation
```

```python
# Useful functions for finding and plotting the locations of nodes

def vectorized_weighted_geographic_center(lat_arr, long_arr, weights):
    lat_arr = np.radians(lat_arr)
    long_arr = np.radians(long_arr)
    x = np.cos(lat_arr) * np.cos(long_arr)
    y = np.cos(lat_arr) * np.sin(long_arr)
    z = np.sin(lat_arr)
    if len(weights.shape) > 1:
        total_weights = np.sum(weights, axis=1)
    else:
        total_weights = np.sum(weights)
    weighted_avg_x = np.sum(weights * x, axis=1) / total_weights
    weighted_avg_y = np.sum(weights * y, axis=1) / total_weights
    weighted_avg_z = np.sum(weights * z, axis=1) / total_weights
    central_longitude = np.arctan2(weighted_avg_y, weighted_avg_x)
    central_sqrt = np.sqrt((weighted_avg_x ** 2) + (weighted_avg_y ** 2))
    central_latitude = np.arctan2(weighted_avg_z, central_sqrt)
    return np.degrees(central_latitude), np.degrees(central_longitude)

def get_locations_time(ts, times, location):
    tables = ts.tables

    time_locations = []
    logtime = np.exp(np.geomspace(np.log(100), np.log(np.max(times)), 384))
    lintime = np.linspace(0, logtime[0], 64)[:-1]

    time_intervals_log = np.concatenate([lintime, logtime])
    ancestor_children = []
    for time in time_intervals_log:
        edges = np.logical_and(
            times[ts.tables.edges.child] <= time,
            times[ts.tables.edges.parent] > time,
        )
        time_slice_child = ts.tables.edges.child[edges]
        time_slice_parent = ts.tables.edges.parent[edges]
        ancestor_children.append(time_slice_child)
        edge_lengths = times[time_slice_parent] - times[time_slice_child]
        weight_parent = 1 - ((times[time_slice_parent] - time) / edge_lengths)
        weight_child = 1 - ((time - times[time_slice_child]) / edge_lengths)
        lat_arr = np.vstack(
            [location[time_slice_parent][:, 0], location[time_slice_child][:, 0]]
        ).T
        long_arr = np.vstack(
            [location[time_slice_parent][:, 1], location[time_slice_child][:, 1]]
        ).T
        weights = np.vstack([weight_parent, weight_child]).T
        lats, longs = vectorized_weighted_geographic_center(
            lat_arr, long_arr, weights
        )
        avg_locations = np.array([lats, longs]).T
        time_locations.append(avg_locations)
    return time_locations, time_intervals_log
```

```python
# First we load a pre-simulated tree sequence
ts = tskit.load("spatial_sim.trees")
```

```python
# Gather the locations of sample nodes (leaves)
sample_locations = []
for node in ts.samples():
    sample_locations.append(ts.individual(node).location)
sample_locations = np.array(sample_locations)[:,:2]
```

```python
# Plot locations of sample nodes
plt.scatter(np.array(sample_locations)[:,0], np.array(sample_locations)[:,1])
```

```python
# Now gather locations of all simulated nodes
locations = []
for node in ts.nodes():
    locations.append(ts.individual(node.individual).location)
locations = np.array(locations)
```

```python
# Now plot the locations of all nodes over time

%matplotlib notebook

time_locations, time_intervals_log = get_locations_time(ts, ts.tables.nodes.time, locations)
fig = plt.figure(figsize=(10, 6))
ax = plt.axes(xlim=(0, 1), ylim=(0, 1))

scat = ax.scatter(
    time_locations[0][:, 0],
    time_locations[0][:, 1],
    s=1,
    alpha=1,
)

def animate(i):
    #############
    # Code to plot individual ancestors
    scat.set_offsets(np.c_[time_locations[i][:, 0], time_locations[i][:, 1]])
    #############
    plt.title(str(int(np.round(time_intervals_log[i] * 25, -2))) + " Years Ago")

anim = FuncAnimation(fig, animate, interval=20, frames=len(time_intervals_log), repeat=True)
```

```python
# Now let's infer locations
# First we need to do a bit of magic with the simulated tree sequence
# Details aren't too important, but basically a forwards time simulation may have marginal trees with multiple roots
# So we run a coalescent (backwards time) simulation so each marginal tree will have a single root
ts = pyslim.load("spatial_sim.trees")
recap_ts = ts.recapitate(recombination_rate=1e-8, Ne=1000).simplify()
```

```python
# Now run the tspyro inference
pyro_time, location, migration_scale, guide = tspyro.infer_geotime(
    recap_ts, torch.tensor(sample_locations), mutation_rate=1e-7)
```

```python
# Finally, we plot both the simulated (real) locations in red and the inferred locations in blue

%matplotlib notebook
inferred_time_locations, time_intervals_log = get_locations_time(recap_ts, recap_ts.tables.nodes.time, location)

together_time_locs = []
for i in zip(time_locations, inferred_time_locations):
    together_time_locs.append(np.concatenate([i[0], i[1]]))

fig = plt.figure(figsize=(10, 6))
ax = plt.axes(xlim=(0, 1), ylim=(0, 1))

colors = np.concatenate([np.repeat("r", together_time_locs[0].shape[0]//2),
                         np.repeat("b", together_time_locs[0].shape[0]//2)])
scat = ax.scatter(
    together_time_locs[0][:, 0],
    together_time_locs[0][:, 1],
    s=1,
    alpha=1,
    c=colors
)

def animate(i):
    #############
    # Code to plot individual ancestors
    scat.set_offsets(np.c_[together_time_locs[i][:, 0], together_time_locs[i][:, 1]])
    colors = np.concatenate([np.repeat("r", together_time_locs[i].shape[0]//2),
                             np.repeat("b", together_time_locs[i].shape[0]//2)])
    scat.set_color(c=colors)
    #############
    plt.title(str(int(np.round(time_intervals_log[i] * 25, -2))) + " Years Ago")

    
anim = FuncAnimation(fig, animate, interval=20, frames=len(time_intervals_log), repeat=True)
```
